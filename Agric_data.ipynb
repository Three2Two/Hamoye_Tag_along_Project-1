```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
url = 'https://raw.githubusercontent.com/ooseni/Hamoye-Datasets/main/FoodBalanceSheets_E_Africa_NOFLAG%20(2).csv' #loading the dataset
agric_data = pd.read_csv(url, sep=',', encoding='latin-1')
agric_data.describe(include='all') #for the mean and std deviation of the whole dataset
```
